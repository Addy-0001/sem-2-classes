```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
category = ["Apple", "Mango", "Orange"]
values = [10, 14, 12]
```

```python
category
```

```python
plt.figure(figsize=(4,4))
plt.bar(category, values, color='skyblue')
plt.show()
```

```python
plt.figure(figsize = (4,4))
plt.pie(values, labels = category, autopct='%1.1f%%', startangle = 90, colors = ["lightblue", "red", 'lightgreen'])
plt.title("Pie Chart of above example")
plt.show()
```

```python
plt.plot(category, values, marker = 's', linestyle = "-")
plt.show()
```
